```python
cate_count_regsum = df.groupby(["Category", "Country/Region"])[["Profit","Sales"]].sum()
cate_count_regsum
```

```python
cate_count_regmean = df.groupby(["Category", "Country/Region"])[["Profit","Sales"]].mean()
cate_count_regmean
```

```python
df['pro/sales'] = df['pro/sales'].round(2)
df['Discount'] = df['Discount'].round(1)
df_sorted = df.sort_values(by=["pro/sales","Discount"], ascending=[False, False])
pro_dis_rel = df_sorted[["pro/sales","Discount"]].drop_duplicates(subset=["pro/sales","Discount"])
print(pro_dis_rel)
```

```python
id_group_combined = df.groupby("Customer ID")[["Sales"]].agg(Sales_sum=('Sales','sum'), Sales_cnt =('Sales','count'))
id_group_combined
```
